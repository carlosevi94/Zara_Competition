```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import os
import seaborn as sns

%matplotlib inline
```

```python
df_pd = pd.read_csv('../bets/week1_bets.csv')
df_pw = pd.read_csv('../bets/week1_bets_pw.csv')
df_pd.head()
```

```python
len(df_pd.block_id.unique()), len(df_pw.block_id.unique())
```

```python
df_pd = df_pd[['block_id', 'gain_per_block', 'n_products']]
df_pd.columns = ['block_id', 'gain_per_block_pd', 'n_products']

df_pw = df_pw[['block_id', 'gain_per_block']]
df_pw.columns = ['block_id', 'gain_per_block_pw']
```

```python
df_pd.head()
```

```python
df_pw.head()
```

```python
df_join = df_pd.join(df_pw.set_index('block_id'), on='block_id')
df_join.head()
```

```python
df_join['mean_gain_per_block'] = (df_join['gain_per_block_pd']+df_join['gain_per_block_pw'])/2
df_join = df_join.sort_values('mean_gain_per_block', ascending=False)
df_join['product_cumsum'] = df_join.n_products.cumsum()
df_join.head(10)
```
